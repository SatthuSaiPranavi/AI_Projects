```python
import re

def chatbot_response(user_input):
    user_input = user_input.lower()

    # Greetings
    if re.search(r"\b(hi|hello|hey|hola)\b", user_input):
        return "Hello! ðŸ˜Š How can I help you today?"

    # Asking name
    elif re.search(r"\b(your name|who are you)\b", user_input):
        return "I'm a simple rule-based chatbot ðŸ¤–, nice to meet you!"

    # Help
    elif re.search(r"\b(help|support|assist)\b", user_input):
        return "Sure! I can chat, answer basic questions, and keep you company."

    # AI related
    elif re.search(r"\b(ai|artificial intelligence|machine learning)\b", user_input):
        return "AI is about making machines think and learn like humans!"

    # Goodbye
    elif re.search(r"\b(bye|exit|quit)\b", user_input):
        return "Goodbye! ðŸ‘‹ Have a great day!"

    # Default response
    else:
        return "Hmm ðŸ¤” I didn't quite understand that. Can you rephrase?"

def start_chatbot():
    print("ðŸ¤– Chatbot: Hello! Type 'bye' to exit.")

    while True:
        user_input = input("You: ")
        response = chatbot_response(user_input)
        print("ðŸ¤– Chatbot:", response)

        if re.search(r"\b(bye|exit|quit)\b", user_input.lower()):
            break

# Run chatbot
start_chatbot()
```

```python
import re

user_name = None

def chatbot_response(user_input):
    global user_name
    user_input = user_input.lower()

    if re.search(r"\bmy name is (\w+)", user_input):
        user_name = re.findall(r"\bmy name is (\w+)", user_input)[0]
        return f"Nice to meet you, {user_name.capitalize()} ðŸ˜Š"

    elif re.search(r"\bhello|hi|hey\b", user_input):
        if user_name:
            return f"Hello {user_name.capitalize()}! How can I help you?"
        return "Hello! ðŸ˜Š What's your name?"

    elif re.search(r"\bhow are you\b", user_input):
        return "I'm doing great! Thanks for asking ðŸ’™"

    elif re.search(r"\bbye|exit\b", user_input):
        return "Bye! ðŸ‘‹ It was nice chatting with you!"

    else:
        return "I'm still learning ðŸ¤–. Try asking something else."

def start_chatbot():
    print("ðŸ¤– Chatbot: Hi! Tell me your name ðŸ˜Š")

    while True:
        user_input = input("You: ")
        response = chatbot_response(user_input)
        print("ðŸ¤– Chatbot:", response)

        if re.search(r"\bbye|exit\b", user_input.lower()):
            break

start_chatbot()
```
